# Access to Green/Blue Areas for Small Areas

Script to assess the amount of green/blue areas within or near each small area.

## Script Setup

```r
suppressPackageStartupMessages(library(sf))
suppressPackageStartupMessages(library(rgdal))
suppressPackageStartupMessages(library(rgeos))
suppressPackageStartupMessages(library(raster))
suppressPackageStartupMessages(library(tidyverse))
suppressPackageStartupMessages(library(geojsonio))
suppressPackageStartupMessages(library(exactextractr))
```

```r
#set a directory for export 
exportDirectory <- "C:/GitHub/Climate-Ireland/Vulnerability/1_InputData/1c_GreenBlue"
```

# Small Areas Data

```r
#get the small area spatial data
smallAreaShp <- st_read("D:/Spatial_Data/CSO/Census_2016_Small_Areas/Shapefiles/Extracted/Small_Areas_Ungeneralised__OSi_National_Statistical_Boundaries__2015/Small_Areas_Ungeneralised__OSi_National_Statistical_Boundaries__2015.shp")
smallAreaShp <- as(smallAreaShp[c('GUID')], "Spatial")

#need to put it into the correct projection
projIreNet95 <- "+proj=tmerc +lat_0=53.5 +lon_0=-8 +k=0.99982 +x_0=600000 +y_0=750000 +ellps=GRS80 +towgs84=0,0,0,0,0,0,0 +units=m +no_defs"
smallAreaShp <- spTransform(smallAreaShp, projIreNet95)

# testData <- smallAreaShp[1:5,]
# plot(testData)
```

## Buffer Each Small Area

Each small area is buffered to assess how much green/blue space is within or close to each small area.

The buffer distance represents walking time - 800 m is approx a 10 minute walking distance.

Could improve by using road/walking routes to calculate travel time.

```r
# https://ecosystems.psu.edu/research/labs/walter-lab/manual/chapter1/1.10-creating-buffers

#set the buffer distance
distance = 800

#10 minute walking distance is about 800 m  http://www.fieldsintrust.org/green-space-index/technical-notes
# smallAreasBuffer <- gBuffer(smallAreaShp, width = distance,byid = TRUE) #byID stops the buffers being dissolved together
# writeOGR(smallAreasBuffer, "./smallAreasBuffer", 'smallAreasBuffer', driver="ESRI Shapefile")

smallAreasBuffer <- st_read('D:/Spatial_Data/CSO/Census_2016_Small_Areas/Shapefiles/Extracted/Small_Areas_Ungeneralised__OSi_National_Statistical_Boundaries__2015/smallAreasBuffer.shp')
smallAreasBuffer <- as(smallAreasBuffer, "Spatial")

#convert to WGS84
projWGS84 <- "+proj=longlat +ellps=WGS84 +datum=WGS84 +no_defs "
smallAreasBufferWGS84 <- spTransform(smallAreasBuffer, projWGS84)
```

# CORINE Green/Blue Data

```r
#gets the CORINE land cover data
corineRaster <- raster('D:/Spatial_Data/Copernicus/LMS/CORINE2018/Extracted/Processed/clc2018_irl_WGS84.tif')

#reclassify CORINE to 1 and 0 so that green/blue areas are 1.
reclassifyValuesCorine <- c(111, 134, 0,
                            141, 524, 1,
                            999, 1000, 0)
rclmatCorine <- matrix(reclassifyValuesCorine, ncol=3, byrow=TRUE)

#reclassify
corine <- reclassify(corineRaster, rclmatCorine, include.lowest = TRUE, right = FALSE)
```

# Analysis

```r
#convert to a simple feature
smallAreasBufferSF <- st_as_sf(smallAreasBufferWGS84)

#calculate the number of cells within each buffered Small Area - uses a simple feature
smallAreasBufferSF$countCORINE <- exact_extract(corine, smallAreasBufferSF, 'count')

#calculate the number of green/blue cells in each small area
smallAreasBufferSF$sumCORINE <- exact_extract(corine, smallAreasBufferSF, 'sum')

#calculate the percentage of each buffered small area that contains green/blue cells
smallAreasBufferSF$greenBluePercentCORINE <- (smallAreasBufferSF$sumCORINE/smallAreasBufferSF$countCORINE)*100

```

```r
#scale the percentages to z scores
smallAreasBufferSF$greenBlue_pct_Z <- scale(smallAreasBufferSF$greenBluePercentCORINE)
head(smallAreasBufferSF)
```

```r
plot(smallAreasBufferSF['greenBluePercentCORINE']) 
```

# Export

```r
#build export path
exportPath <- paste(exportDirectory, 'accessToGreenBlueAreas', sep = '/')
exportPath <- paste(exportPath, '.csv', sep = '')

# st_write(smallAreasBufferSF['GUID', 'greenBluePercentCORINE'], exportPath, driver = "CSV")
smallAreasBufferExport <- as(smallAreasBufferSF, "Spatial")

smallAreasBufferExport <- as.data.frame(smallAreasBufferExport) %>% select(1,4,5)

write.csv(smallAreasBufferExport, exportPath, row.names = FALSE)
```
